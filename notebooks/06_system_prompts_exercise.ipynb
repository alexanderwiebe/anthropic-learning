```python
from dotenv import load_dotenv

load_dotenv()

from anthropic import Anthropic
from anthropic.types import MessageParam

client = Anthropic()
model = "claude-sonnet-4-6"


def add_user_message(messages: list[MessageParam], text: str) -> None:
    user_message: MessageParam = {"role": "user", "content": text}
    messages.append(user_message)


def add_assistant_message(messages: list[MessageParam], text: str) -> None:
    assistant_message: MessageParam = {"role": "assistant", "content": text}
    messages.append(assistant_message)


def chat(messages: list[MessageParam], system: str | None = None) -> str:
    if system:
        message = client.messages.create(
            model=model,
            max_tokens=1000,
            messages=messages,
            system=system,
        )
    else:
        message = client.messages.create(
            model=model,
            max_tokens=1000,
            messages=messages,
        )
    assert message.content[0].type == "text"
    return message.content[0].text
```

```python
from anthropic.types import MessageParam

messages: list[MessageParam] = []

system_prompt = """
you are a developer that is returning only the coded answer, no explanations,
no comments, just the code.
I want the code to be readable and well formatted but should be short and concise.
"""
add_user_message(
    messages, "write a python function that checks a string for duplicate characters"
)
answer = chat(messages, system=system_prompt)

answer
```
